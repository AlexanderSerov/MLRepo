```python
seed = 1
```

```python
image_generator = TrainGen.flow_from_directory(
    'C:/Users/cah4e/Desktop/BottomDataset/train/AugmentedDataset/images/',
    target_size=(500,500),
    class_mode=None,
    # batch_size=5,
    save_to_dir='images',
    save_prefix='aug',
    save_format='png',
    seed=seed)

mask_generator = MaskGen.flow_from_directory(
    'C:/Users/cah4e/Desktop/BottomDataset/train/AugmentedDataset/masks/',
    target_size=(500,500),
    color_mode = 'grayscale',
    class_mode=None,
    # batch_size=5,
    save_to_dir='masks',
    save_prefix='aug',
    save_format='png',
    seed=seed)

train_generator = zip(image_generator, mask_generator)
```
