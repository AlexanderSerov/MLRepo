```python
%matplotlib notebook

# Import libs
import matplotlib
import matplotlib.axes
import matplotlib.figure
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Perceptron
class PerceptronNN:
    
    def __init__(self, NumberOfInputs):
        
        self.weights = np.random.randn(1,NumberOfInputs)
        
    def FeedForward(self, Dot):
        return ActivationFunction(Dot.inputs @ self.weights.T)
    
    def Train(self, Dot):
        
        guess = self.FeedForward(Dot)
        error = Dot.label - guess
        self.weights += error * Dot.inputs * 0.1
        
    def GetWeights(self):
        return self.weights
```
